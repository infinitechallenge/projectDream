# Print 0~100, 0~200, 0~300 in different arrays 

```python
x=[]
for i in range(100):
    x.append(i)
print(x)
```

```python
#start the loop that will loop 3 times.
#make an empty array and call it an X
#First, 0 from the range(100), do 0*0 and append it to x. Now you have [0]
#Second, 1 from the range(100), do 1*0, and append it to x, Now you have [0,0] 
#do this 100 times and you have an array [0,...,0]
#Now go to the top forloop, your i is 1 now. 
#Create an empty array and call it an X
#First, 0 from the range(100), do 0*1 and append it to x. Now you have [0]
#Second, 1 from the range(100), do 1*1, and append it to x, Now you have [0,1] 
#do this 100 times and you have an array [0,1,2...,99]
for i in range(4):
    x=[]
    for j in range(100):
        x.append(j*i)
    print(x)
    
```

# Make the other half triangle with *

```python
#if range is negative, it won't append anything
for i in range(4):
    z=[]
    for j in range(4-i):
        z.append(' ')
    for j in range(i+1):
        z.append('*')
    print(z)
```
